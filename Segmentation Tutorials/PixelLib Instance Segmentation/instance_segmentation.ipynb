```python
# https://github.com/matterport/Mask_RCNN/releases
# https://github.com/matterport/Mask_RCNN/releases/tag/v2.0

segmentation_model = instance_segmentation()
segmentation_model.load_model('/content/mask_rcnn_coco.h5')
```

```python
from matplotlib import pyplot as plt 
```

```python
# Apply instance segmentation
image = cv2.imread('/content/horse.jpg')

res = segmentation_model.segmentFrame(image, show_bboxes=True)
image = res[1]

plt.rcParams["figure.figsize"] = [10, 10]
plt.imshow(image, cmap='gray')
plt.show()
```
